```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r"iphones.csv")
```

```python

# Group the data by color and calculate the average price for each color
avg_price_by_color = data.groupby('Color')['Price'].mean().sort_values(ascending=False)

# Plot the results
plt.figure(figsize=(10,6))
sns.barplot(x=avg_price_by_color.index, y=avg_price_by_color.values, palette='viridis')
plt.xticks(rotation=90)
plt.title('Average Price by Color')
plt.xlabel('Color')
plt.ylabel('Average Price')
plt.show()
```
